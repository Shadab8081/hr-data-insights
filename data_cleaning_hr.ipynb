```python
#Import Libraries
import pandas as pd
import numpy as np
```

# Import CSV Files

```python
df1 = pd.read_csv('/content/employee_data.csv')
df2 = pd.read_csv('/content/employee_engagement_survey_data.csv')
df3 = pd.read_csv('/content/recruitment_data.csv')
df4 = pd.read_csv('/content/training_and_development_data.csv')
```

# Renaming Columns Specifically for Merging

```python
df1_rename = df1
```

```python
df2_rename = df2.rename(columns = {'Employee ID' : 'EmpID'})
```

```python
df3_rename = df3.rename(columns = {'Applicant ID' : 'EmpID'})
```

```python
df4_rename = df4.rename(columns = {'Employee ID' : 'EmpID'})
```

```python
merged_df = df1_rename.merge(df2_rename, on='EmpID') \
               .merge(df3_rename, on='EmpID') \
               .merge(df4_rename, on='EmpID')
```
